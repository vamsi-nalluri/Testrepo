```python
import pandas as pd
import seaborn as s
import matplotlib.pyplot as plt
vs['decade'] = (vs['year'] // 10) * 10
decade_film_counts = vs.groupby('decade')['title'].count().reset_index()
s.barplot(data=decade_film_counts, x='decade', y='title', palette='viridis')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=50)
plt.show()
```
