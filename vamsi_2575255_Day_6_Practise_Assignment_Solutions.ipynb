```python
import datetime  # Import the datetime module

class Question:
    def __init__(self, text, options, correct_option):
        self.text = text
        self.options = options
        self.correct_option = correct_option

    def is_correct(self, user_answer):
        return user_answer == self.correct_option

class Quiz:
    def __init__(self, name, questions):
        self.name = name
        self.questions = questions

    def take_quiz(self):
        score = 0
        for question in self.questions:
            print(question.text)
            for i, option in enumerate(question.options, 1):
                print(f"{i}. {option}")

            user_answer = int(input("Your answer (enter the option number): "))
            if question.is_correct(user_answer):
                print("Correct!\n")
                score += 1
            else:
                print(f"Wrong! The correct answer was option {question.correct_option}.\n")

        print(f"Quiz completed! Your score: {score}/{len(self.questions)}")
        return score

# Example usage:
# Create questions
question1 = Question("What is the capital of France?", ["Paris", "London", "Berlin", "Madrid"], 1)
question2 = Question("Which planet is known as the Red Planet?", ["Mars", "Venus", "Jupiter", "Saturn"], 1)

# Create a quiz
quiz1 = Quiz("General Knowledge Quiz", [question1, question2])

# Take the quiz
user_score = quiz1.take_quiz()

# Store user progress, history, and timestamp using lists
user_progress = {
    "quiz_name": quiz1.name,
    "score": user_score,
    "timestamp": datetime.datetime.now()  # Get the current timestamp
}
user_history = []  # List to store user history

# Append user progress to history
user_history.append(user_progress)

# Example: Retrieve and display user progress, history, and timestamps
for progress in user_history:
    print(f"\nQuiz Name: {progress['quiz_name']}, Occurred Score: {progress['score']}/{len(quiz1.questions)} Timestamp: {progress['timestamp']}")
    
```

# 3. You are tasked with designing a Python class hierarchy for a simple banking system. The system should be able to handle different types of accounts, such as Savings Accounts and Checking Accounts. Both account types should have common attributes like an account number, account holder's name, and balance. However, Savings Accounts should have an additional attribute for interest rate, while Checking Accounts should have an attribute for overdraft limit.

    1. Create a Python class called BankAccount with the following attributes and methods:
        
        a. Attributes: account number, account holder_name, balance
            
        b. Methods: init_() (constructor), deposit(), and withdraw

    2. Create two subclasses. Savings Account and Checking Account, that inherit from the BankAccount class.

    3. Add the following attributes and methods to each subclass:

        a. Savings Account:

            i. Additional attribute: interest rate

            ii. Method: calculate interest(), which calculates and adds interest to the account based on the interest rate.

        b. Checking Account:

    4. Write a program that creates instances of both Savings Account and Checking Account and demonstrates the use of their methods.

    7. Provide comments in your code to explain the purpose of each class, attribute, and method.

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.__account_number = account_number  # Private attribute
        self.__account_holder_name = account_holder_name  # Private attribute
        self.__balance = balance  # Private attribute

    # Getter method
    def get_account_number(self):
        return self.__account_number

    def get_account_holder_name(self):
        return self.__account_holder_name

    def get_balance(self):
        return self.__balance

    # Setter method
    def set_balance(self, balance):
        if balance < 0:
            raise ValueError("Balance cannot be negative")
        self.__balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be greater than zero")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be greater than zero")

        if amount > self.__balance:
            raise ValueError("Insufficient balance")

        self.__balance -= amount
        return amount


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, interest_rate=0.0):
        super().__init__(account_number, account_holder_name, balance)  # Call parent class constructor
        self.__interest_rate = interest_rate  # Private attribute

    # Getter and setter methods 
    def get_interest_rate(self):
        return self.__interest_rate

    def set_interest_rate(self, interest_rate):
        if interest_rate < 0:
            raise ValueError("Interest rate cannot be negative")
        self.__interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.get_balance() * (self.__interest_rate / 100)  # Use getter method to access balance
        self.set_balance(self.get_balance() + interest)  # Use setter method to update balance


class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, overdraft_limit=0.0):
        super().__init__(account_number, account_holder_name, balance)  # Call parent class constructor
        self.__overdraft_limit = overdraft_limit  # Private attribute

    # Getter and setter methods 
    def get_overdraft_limit(self):
        return self.__overdraft_limit

    def set_overdraft_limit(self, overdraft_limit):
        if overdraft_limit < 0:
            raise ValueError("Overdraft limit should be greater than zero")
        self.__overdraft_limit = overdraft_limit

    def withdraw(self, amount):

        if amount <= 0:
            raise ValueError("Withdrawal amount Should not be less than one")

        if amount > (self.get_balance() + self.__overdraft_limit):  # Use getter method to access balance
            raise ValueError("Withdrawal amount exceeded")

        if amount > self.get_balance():  # Use getter method to access balance
            # Deduct the remaining amount from overdraft limit
            self.__overdraft_limit -= (amount - self.get_balance())  # Use getter method to access balance
            self.set_balance(0)  # Use setter method to update balance
            return amount
        else:
            self.set_balance(self.get_balance() - amount)  # Use setter method to update balance
            return amount



# program to demonstrate the use of classes
if __name__ == "__main__":
    # Create a Savings Account
    savings_account = SavingsAccount("SA123456", "John Doe", 1000.0, 3.5)
    print("\nSavings Account Balance :", savings_account.get_balance())
    savings_account.calculate_interest()
    print("\nSavings Account Balance after interest :", savings_account.get_balance())

    # Create a Checking Account
    checking_account = CheckingAccount("CA789012", "Jane Smith", 500.0, 100.0)
    print("\nChecking Account Balance :", checking_account.get_balance())

    # Deposit and withdraw from the accounts
    savings_account.deposit(200.0)
    print("\nSavings Account Balance after deposit :", savings_account.get_balance())

    try:
        checking_account.withdraw(700.0)
        print("\nChecking Account Balance after withdrawal :", checking_account.get_balance())
    except ValueError as e:
        print("\nError:", e)
```

# 5. Library Management System-Scenario: You are developing a library management system where you need to handle books, patrons, and library transactions.

    1. Create a class hierarchy that includes classes for books (e.g., Book), patrons (e.g.. Patron), and transactions (e.g.. Transaction). Define attributes and methods for each class.

    2. Implement encapsulation by making relevant attributes private and providing getter and setter methods where necessary.

    3. Use inheritance to represent different types of books (e.g., fiction, non-fiction) as subclasses of the Book class. Ensure that each book type can have specific attributes and methods.

    4. Demonstrate polymorphism by allowing patrons to check out and return books, regardless of the book type.

    5. Implement a method for tracking overdue books and notifying patrons.

    6. Consider scenarios like book reservations, late fees, and library staff interactions in your design.

```python
import datetime

# Base class for all types of books
class Book:
    def __init__(self, bid, title, author, book_type):
        self.__bid = bid
        self.title = title
        self.author = author
        self.book_type = book_type
        self.status = True  # True indicates the book is available
        self.due = None  # Due date for the book

    # Getter method for book ID
    def get_bid(self):
        return self.__bid

    # Getter method for book type
    def get_book_type(self):
        return self.book_type

    # Setter method for book status
    def set_status(self, status):
        self.status = status

    # Setter method for due date
    def set_due(self, due):
        self.due = due

# Subclass for Fiction books
class Fiction(Book):
    def __init__(self, bid, title, author):
        super().__init__(bid, title, author, "Fiction")

# Subclass for Non-Fiction books
class NonFiction(Book):
    def __init__(self, bid, title, author):
        super().__init__(bid, title, author, "Non-Fiction")

# Patron class to represent library patrons
class Patron:
    def __init__(self, pid, name, phone):
        self.pid = pid
        self.name = name
        self.phone = phone
        self.checked_out_books = []

    # Method for checking out a book
    def check_out(self, book, due_date):
        if book.status:
            book.set_status(False)
            book.set_due(due_date)
            self.checked_out_books.append(book)
            print(f"\nChecked out '{book.title}' by {book.author} due on {due_date}")
        else:
            raise Exception("\nBook is not available for checkout.")

    # Method for returning a book
    def return_book(self, book):
        if book in self.checked_out_books:
            book.set_status(True)
            book.set_due(None)
            self.checked_out_books.remove(book)
            print(f"\nReturned '{book.title}' by {book.author}")
        else:
            raise Exception("\nThis book is not checked out by this patron.")

# Transaction class to represent library transactions
class Transaction:
    def __init__(self, transaction_id, patron, book, due_date):
        self.transaction_id = transaction_id
        self.patron = patron
        self.book = book
        self.due_date = due_date

# Library class to manage books, patrons, and transactions
class Library:
    def __init__(self):
        self.books = []
        self.patrons = []
        self.transactions = []
        self.yearly_fee = 5  # Example yearly late fee

    # Method to track overdue books
    def track_overdue_books(self):
        today = datetime.date.today()
        overdue_books = []

        for patron in self.patrons:
            for book in patron.checked_out_books:
                if book.due and book.due < today:
                    overdue_books.append((patron, book))

        return overdue_books

    # Method to notify overdue patrons and calculate late fees
    def notify_overdue_patrons(self):
        overdue_books = self.track_overdue_books()
        for patron, book in overdue_books:
            late_fee = self.calculate_late_fee(book.due)
            # Notify the patron about the overdue book and late fee
            print(f"\nPatron {patron.name} has an overdue book: '{book.title}' by {book.author}. Late Fee: ${late_fee}")

    # Method to calculate late fee based on days overdue
    def calculate_late_fee(self, due_date):
        today = datetime.date.today()
        days_overdue = (today - due_date).days
        return days_overdue * self.yearly_fee

# Example usage:

# Create some book instances
bk1 = Fiction(bid=1, title="The Sun", author="CBN")
bk2 = NonFiction(bid=2, title="The Moon", author="Jin")

# Create a library instance
library = Library()

# Add books to the library
library.books.extend([bk1, bk2])

# Create a patron
ptr1 = Patron(pid=1, name="Tinku", phone="128749562")
library.patrons.append(ptr1)

# Try checking out a book
try:
    ptr1.check_out(bk1, due_date=datetime.date(2023, 9, 30))
except Exception as e:
    print(e)

# Try checking out the same book again (should fail)
try:
    ptr1.check_out(bk1, due_date=datetime.date(2023, 9, 30))
except Exception as e:
    print(e)

# Try returning a book
try:
    ptr1.return_book(bk1)
except Exception as e:
    print(e)

# Try returning the same book again (should fail)
try:
    ptr1.return_book(bk1)
except Exception as e:
    print(e)

# Notify overdue patrons about late fees
library.notify_overdue_patrons()
```

# 6.Online Shopping Cart

Scenario: You are tasked with designing a class hierarchy for an online shopping cart system. The system should handle products, shopping carts, and orders. Consider various OOP principles while designing this system.

    1. Create a class hierarchy that includes classes for products (e.g., Product), shopping carts (e.g., ShoppingCart), and orders (e.g., Order). Define attributes and methods for each class.

    2. Implement encapsulation by making relevant attributes private and providing getter and setter methods where necessary.

    3. Use inheritance to represent different types of products (e.g., electronics, clothing) as subclasses of the Product class. Ensure that each product type can have specific attributes and methods.

    4. Demonstrate polymorphism by allowing various product types to be added to a shoppingcart and calculate the total cost of items in the cart.

    5. Implement a method for placing an order, which transfers items from the shopping cart to an order.Consider scenarios like out-of-stock products, discounts, and shipping costs in your design.

```python
# Online Shopping Cart System

# Product class represents a generic product
class Product:
    def __init__(self, product_id, name, price, stock):
        self.__product_id = product_id
        self.__name = name
        self.__price = price
        self.__stock = stock

    # Getter methods
    def get_product_id(self):
        return self.__product_id

    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    def get_stock(self):
        return self.__stock

    # Setter method for stock
    def set_stock(self, stock):
        self.__stock = stock

    # Check if the product is available
    def is_available(self):
        return self.__stock > 0

    # Calculate discounted price
    def calculate_discounted_price(self, discount_percentage):
        return self.__price * (1 - discount_percentage / 100)


# Electronics class represents a type of product with additional attributes
class Electronics(Product):
    def __init__(self, product_id, name, price, stock, warranty_period):
        super().__init__(product_id, name, price, stock)
        self.__warranty_period = warranty_period

    # Getter method for warranty period
    def get_warranty_period(self):
        return self.__warranty_period


# Clothing class represents a type of product with additional attributes
class Clothing(Product):
    def __init__(self, product_id, name, price, stock, size):
        super().__init__(product_id, name, price, stock)
        self.__size = size

    # Getter method for size
    def get_size(self):
        return self.__size


# ShoppingCart class represents a shopping cart
class ShoppingCart:
    def __init__(self):
        self.__items = []

    # Getter method for cart items
    def get_cart_items(self):
        return self.__items

    # Add an item to the cart
    def add_item(self, product, quantity=1):
        if product.is_available() and quantity > 0:
            for item in self.__items:
                if item["product"] == product:
                    item["quantity"] += quantity
                    break
            else:
                self.__items.append({"product": product, "quantity": quantity})

    # Remove an item from the cart
    def remove_item(self, product, quantity=1):
        for item in self.__items:
            if item["product"] == product:
                if item["quantity"] > quantity:
                    item["quantity"] -= quantity
                else:
                    self.__items.remove(item)

    # Calculate the total cost of items in the cart
    def calculate_total_cost(self):
        total_cost = 0
        for item in self.__items:
            total_cost += item["product"].get_price() * item["quantity"]
        return total_cost


# Order class represents an order with items from the shopping cart
class Order:
    def __init__(self, order_id, customer, items, shipping_cost=0):
        self.__order_id = order_id
        self.__customer = customer
        self.__items = items
        self.__shipping_cost = shipping_cost

    # Getter methods for order details
    def get_order_id(self):
        return self.__order_id

    def get_customer(self):
        return self.__customer

    def get_items(self):
        return self.__items

    def get_shipping_cost(self):
        return self.__shipping_cost

    # Calculate the total cost of the order
    def calculate_total_cost(self):
        total_cost = 0
        for item in self.__items:
            total_cost += item["product"].get_price() * item["quantity"]
        total_cost += self.__shipping_cost
        return total_cost


# Example usage:
if __name__ == "__main__":
    # Create some product instances
    laptop = Electronics(1, "Laptop", 800, 10, "2 years")
    t_shirt = Clothing(2, "T-shirt", 20, 50, "Large")

    # Create a shopping cart
    cart = ShoppingCart()

    # Add items to the shopping cart
    cart.add_item(laptop, 2)
    cart.add_item(t_shirt, 3)

    # Calculate the total cost in the shopping cart
    total_cost_in_cart = cart.calculate_total_cost()
    print(f"Total cost in the shopping cart: ${total_cost_in_cart}")

    # Checkout and create an order
    order = Order(1, "John Doe", cart.get_cart_items(), shipping_cost=10)

    # Display order details
    print(f"Order ID: {order.get_order_id()}")
    print(f"Customer: {order.get_customer()}")
    print("Ordered Items:")
    for item in order.get_items():
        product = item["product"]
        quantity = item["quantity"]
        print(f"Product: {product.get_name()}, Quantity: {quantity}")
    print(f"Shipping Cost: ${order.get_shipping_cost()}")
    print(f"Total Cost: ${order.calculate_total_cost()}")
```
