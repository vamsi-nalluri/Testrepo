# 1.Write a Python program to find a target values in a list using linear search with following steps:

        a. Initialize the list to store the input elements.

        b. Initialize found-False.

        c. Enter the item to be searched (match_item).
        
        d. For each element in the list

                1. if match item = value

                2. return match item's position.

        e. If the match item is not in the list, display an error message that the item is not found in the list.

```python
def ls(l,v):
    i=0
    found=False     #b
    while i <len(l) and not found:  #d
        if l[i]==v:        #d1
            found=True
        else:
            i+=1
    if found:
        print('\nElement %d Found at Position %d'%(v,i+1))      #d2

    else:
        print('\nElement %d not found in the list'%v)    #e
data=[] #a
n=int(input('\nEnter Size of elements to be inserted:\t'))
for i in range(0,n):
    x=int(input('\nEnter the Element %d:'%(i+1)))
    data.append(x)
    
e=int(input('\nEnter the value to be searched: \t'))   #c
ls(data,e)
```

# 2. Write a Python program to implement binary search to find the target values from the list:

        a. Create a separate function to do binary search.

        b. Get the number of inputs from the user.

        C. Store the inputs individually in a list.

        d. In binary search function at first sort the list in order to start the search from middle of the list.

        e. Compare the middle element to right and left elements to search target element. 
        
        f. If greater, move to right of list or else move to another side of the list.

        g. Print the result along with the position of the element.


```python
def binarysearch(lst,v):  #a
    first=0
    last=len(lst)-1
    found=False
    while(first<=last and not found):
        mid=(first+last)//2   #d
        if lst[mid]==v:
            found=True
        else:
            if v<lst[mid]:                 #e
                last=mid-1         #f
            else:                 #f
                first=mid+1
    if found:
        print('Element %d Found at Position %d'%(v,mid+1))    #g
    else:
        print('Element %d is not found in the list'%(v))

    
data=[]
n=int(input('Enter No of elements to be inserted:\t'))  #b
for i in range(0,n):                                               #c
    x=int(input('Enter the Element %d:'%(i+1)))
    data.append(x)
    
e=int(input('Enter the value to be searched:\t'))
if sorted(data)==data:
    binarysearch(data,e)   
else:
    print("Binary search is only for sorted list")
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm:

        a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

        b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted list.

        c. In the second iteration, find the smallest element in the unsorted list and place it in the correct position by comparing with the element in the sorted list.

        d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list.

        e. This process continues till the unsorted list becomes empty.

        f. Display the sorted list.


```python
def selsort(lst):
    srt=[]    #a
    while lst:  # b,c,d,e
        m=min(lst)
        srt.append(m)  #b
        lst.remove(m)   #b
    return srt
unsl=[]#a
n=int(input("Enter no of elements want to insert in list: "))
for i in range(n):
    v=int(input(f"\nEnter {i}th element: "))
    unsl.append(v)
srt=selsort(unsl)
print("\nAfter Selection sort list : ",srt)  #f
```

# 6. Write a Python script to perform the following operations on a singly linked list

        a. Create a list

        b. Find the smallest element from the list

        c. Insert an element if it is not a duplicate element

        d. Display the elements in reverse order



```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.hd = None

    def is_empty(self):
        return self.hd is None

    def insert(self, data):
        if not self.contains(data):
            newnd = Node(data)
            newnd.next = self.hd
            self.hd = newnd

    def contains(self, data):
        cur = self.hd
        while cur:
            if cur.data == data:
                return True
            cur = cur.next
        return False

    def finds(self):
        if self.is_empty():
            return None
        cur = self.hd
        small = cur.data
        while cur:
            if cur.data < small:
                small = cur.data
            cur = cur.next
        return small

    def display(self):
        stack = []
        cur = self.hd
        while cur:
            stack.append(cur.data)
            cur = cur.next
        while stack:
            print(stack.pop(), end=' ')
vamsi= LinkedList()
n = int(input("\nEnter the number of elements to insert: "))
for i in range(n):
    data = int(input(f"\nEnter element {i + 1}: "))
    vamsi.insert(data)
small = vamsi.finds()
if small is not None:
    print(f"\nThe smallest element in the list is {small}")
else:
    print("\nThe list is empty.")
print("\nElements in reverse order:")
vamsi.display()
```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack

            a. A/B^C+D*E-F*G

            b. (B^2-4*A*C)^(1/2)(100)

```python
# n='A/B^C+D*E-F*G'
# for i in n:
#     print(i)
```

```python
class Stack:
    def __init__(self):
        self.__stack = []

    def size(self):
        return len(self.__stack)

    def is_empty(self):
        if self.size() == 0:
            return True
        else:
            return False

    def push(self, itm):
        self.__stack.append(itm)

    def pop(self):
        if self.is_empty():
            raise Exception("Stack is already empty; we cannot perform POP")
        else:
            return self.__stack.pop()

    def top(self):
        if self.is_empty():
            raise Exception("\nStack underflow")
        else:
            return self.__stack[self.size() - 1]

    def disp(self):
        print(self.__stack)


def pre(op):
    if op.isdigit():
        return 0
    elif op in ('+', '-'):
        return 1
    elif op in ('*', '/'):
        return 2
    elif op == '^':
        return 3
    return -1


def post(n):
    l = []
    stack = Stack()
    for i in n:
        if i.isalpha():
            l.append(i)
        elif i == '(':
            stack.push(i)
        elif i == ')':
            while not stack.is_empty() and stack.top() != '(':
                l.append(stack.pop())
            if not stack.is_empty() and stack.top() == '(':
                stack.pop()
        else:
            while (not stack.is_empty() and stack.top() != '(' and pre(i) <= pre(stack.top())):
                l.append(stack.pop())
            stack.push(i)
    while not stack.is_empty():
        l.append(stack.pop())
    return "".join(l)

if __name__ == '__main__':
    n = 'A/B^C+D*E-F*G'
    n1 = '(B^2-4*A*C)^(1/2)(100)'

    v2=post(n)
    v3=post(n1)
    print(f'Infix expression "{n}" after changing into prefix expression is "{v2}"')
    print(f'\nInfix expression "{n1}" after changing into prefix expression is "{v3}"')
```
